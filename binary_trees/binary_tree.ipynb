# Playing around with binary trees

```python
# Node structure

class Node:
	def __init__ (self, key):
		self.value = key
		self.right = None
		self.left = None
```

Creating a Binary Tree

In this example, we will explore how to create a Binary Tree with four nodes 2, 3, 4, 5.


![Description of image](binary_tree_example1.png)

```python
# initialise nodes and allocate memory
firstNode = Node(2)
secondNode = Node (3)
thirdNode = Node(4)
forthNode = Node(5)

# Connect node

firstNode.left = secondNode
firstNode.right = thirdNode
secondNode.left = forthNode
```
